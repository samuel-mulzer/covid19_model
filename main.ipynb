```python
import covid19_model
simulate = covid19_model.simulate
```

```python
import numpy as np
import matplotlib.pyplot as plt
solve = covid19_model.solver.solve
sir = covid19_model.models.sir

model = sir

t = 0
d = 100
h = 0.01
tspan = np.arange(t, t+d, h)

labels = ['S', 'I', 'R']
s0, i0, r0 = 0.999, 0.001, 0
n = 1
x0 = np.array([s0, i0, r0])

beta = 0.5
gamma = 0.2
args = (beta, gamma)

solution = solve(model, x0, tspan, h, args)
plt.figure(figsize=(8,4))
plt.xlabel('Zeit in Tagen')
plt.ylabel('Anteil an  der Bevlkerung')
c = 0
for x in solution:
    plt.plot(tspan, x, label=labels[c])
    c += 1
plt.legend()
plt.savefig('figures/sir.pdf')
plt.show()
```
